#Finding Z-Score


```python
from scipy.stats import zscore

# Calculate Z-scores for the Quantity column
outlier_scores = zscore(df['Quantity'])

print(outlier_scores)
```

```python
# Create a logical vector indicating whether the outlier score is greater than 3 or less than -3
is_outlier = (outlier_scores > 3) | (outlier_scores < -3)

# Add a column to the DataFrame indicating if Quantity is an outlier
df['is_outlier'] = is_outlier
```

```python
import pandas as pd
import numpy as np
import networkx as nx

test_data = pd.DataFrame({
    'Timestamp': ['2024-12-20 10:00:00', '2024-12-20 12:00:00', '2024-12-20 14:00:00'],
    'From': ['A', 'B', 'A'],
    'To': ['A', 'C', 'A'],
    'Quantity': [50, 200, 600]
})

expected_output = {
    'Year': [2024, 2024, 2024],
    'Month': [12, 12, 12],
    'Hour': [10, 12, 14],
    'Time_Diff': [0.0, 7200.0, 7200.0],
    'Quantity_Log': [3.912023, 5.298317, 6.396930],
    'Quantity_Binned': ['0-100', '101-500', '501-1000'],
    'Is_Outlier': [False, False, False],
    'Is_Self_Trade': [True, False, True],
    'Graph_Nodes': 3,
    'Graph_Edges': 2
}

def test_feature_engineering():
    # Input Data
    data = test_data.copy()

    # Timestamp Features
    #ensuring Extracting  year, month, and hour from the 'Timestamp' column.
    data['Timestamp'] = pd.to_datetime(data['Timestamp'])
    data['Year'] = data['Timestamp'].dt.year
    data['Month'] = data['Timestamp'].dt.month
    data['Hour'] = data['Timestamp'].dt.hour

    # Time Difference
    data['Time_Diff'] = data['Timestamp'].diff().dt.total_seconds().fillna(0)

    # Quantity Log Transformation
    data['Quantity_Log'] = np.log(data['Quantity'])

    # Quantity Binning
    bins = [0, 100, 500, 1000, float('inf')]
    labels = ['0-100', '101-500', '501-1000', '1000+']
    data['Quantity_Binned'] = pd.cut(data['Quantity'], bins=bins, labels=labels, right=False)

    # Outlier Detection
    threshold = 1000
    data['Is_Outlier'] = data['Quantity'] > threshold

    # Self-Trade Detection
    data['Is_Self_Trade'] = data['From'] == data['To']

    # Graph Creation
    G = nx.from_pandas_edgelist(data, 'From', 'To', create_using=nx.DiGraph())

    # Assertions
    assert data['Year'].tolist() == expected_output['Year']
    assert data['Month'].tolist() == expected_output['Month']
    assert data['Hour'].tolist() == expected_output['Hour']
    assert data['Time_Diff'].tolist() == expected_output['Time_Diff']
    np.testing.assert_array_almost_equal(data['Quantity_Log'].values, expected_output['Quantity_Log'], decimal=6)
    assert data['Quantity_Binned'].tolist() == expected_output['Quantity_Binned']
    assert data['Is_Outlier'].tolist() == expected_output['Is_Outlier']
    assert data['Is_Self_Trade'].tolist() == expected_output['Is_Self_Trade']
    assert G.number_of_nodes() == expected_output['Graph_Nodes']
    assert G.number_of_edges() == expected_output['Graph_Edges']

    # Success Message
    print("All feature engineering tests passed successfully!")

# Run the test
test_feature_engineering()
```
